# Míminos, Máximos e Médios


Média de Matemática

```python
media_matematica = df['math score'].mean()
media_matematica
```

Máximo de Matemática

```python
max_matematica = df['math score'].max()
max_matematica
```

Máximo de Leitura

# Média de Matemática por Etnia

```python
media_matematica_por_etnia = df.groupby("race/ethnicity")[['math score']].mean()
media_matematica_por_etnia
```

# Média de Escrita por Escolaridade dos Pais

```python
escolaridade_pais = df.groupby("parental level of education")[['writing score']].mean()
escolaridade_pais
```

# Média de Leitura por Particiação em Curso Preparatório

```python
leitura_preparacao = df.groupby("test preparation course")[['reading score']].mean()
leitura_preparacao
```
